```python
import pandas as pd
import numpy as np
```

```python
Chg=df["Open"]-df["Close"]
df["change_in_price"]=Chg
```

```python
df['change_in_price']=df['Close'].diff()
df.head(50)
```

```python
df.head(1026)
```

```python
df.drop('Trades',
  axis='columns', inplace=True)
df.head(30)
```

```python
mask=df['Symbol']!=df['Symbol'].shift(1)
df['change_in_price']=np.where(mask== True, np.nan,df['change_in_price'])
df.head(1026)
```

```python
df.drop('change_in',
  axis='columns', inplace=True)
df.head(30)
```

```python
n=14
up_df, down_df=df[["Symbol","change_in_price"]].copy(), df[["Symbol","change_in_price"]].copy()
#for up days , if change is less than 0 set to 0.
up_df.loc['change_in_price']=up_df.loc[(up_df['change_in_price'] < 0),'change_in_price']=0
#for down days, if the change is greater than 0 set to 0
down_df.loc['change_in_price']=down_df.loc[(down_df['change_in_price'] > 0), 'change_in_price']=0
down_df['change_in_price']=down_df['change_in_price'].abs()

#calculate the EWMA (exponential weighted moving average), meaning older values are given less weight compared to newer value 
ewma_up=up_df.groupby('Symbol')['change_in_price'].transform(lambda X: X.ewm(span = n).mean())
ewma_down = down_df.groupby('Symbol')['change_in_price'].transform(lambda x: x.ewm(span = n).mean())

#calculate the relative strength
relative_strength = ewma_up / ewma_down

# calculate the relative strength index 
relative_strength_index = 100.00 - (100.00 / (1.0 + relative_strength))

# add the info to the data frame 
df['down_days'] = down_df['change_in_price']
df['up_days'] = up_df['change_in_price']
df['RSI'] = relative_strength_index
```

```python
df.head(30)
```

```python
# calculate stochastic Oscillator
n=14 

# make a copy of the high and low column.
l14,h14=df[['Symbol','Low']].copy(), df[['Symbol','High']].copy()

# group by symbol, then apply the rolling function and grab the min and max.
l14=l14.groupby('Symbol')['Low'].transform(lambda x: x.rolling(window = n).min())
h14=h14.groupby('Symbol')['High'].transform(lambda x: x.rolling(window = n).max())

# calculate the stochastic oscillator 
k_percent=100*((df['Close']-l14)/(h14 - l14))

# add the info to the data frame 
df['low_14']=l14
df['high_14']=h14
df['k_percent']=k_percent
```

```python
df.head(30)
```

```python
# calculate Williams %R
n=14 

# make a copy of the high and low column.
l14,h14=df[['Symbol','Low']].copy(), df[['Symbol','High']].copy()

# group by symbol, then apply the rolling function and grab the min and max.
l14=l14.groupby('Symbol')['Low'].transform(lambda x: x.rolling(window = n).min())
h14=h14.groupby('Symbol')['High'].transform(lambda x: x.rolling(window = n).max())

#calculate %R
r_percent =((h14 - df['Close']) / (h14 - l14))* - 100

#add in data frame 
df['r_percent']= r_percent
```

```python
df.head(30)
```

```python
# calculate the MACD 
ema_26 = df.groupby('Symbol')['Close'].transform(lambda x: x.ewm(span = 26).mean())
ema_12 = df.groupby('Symbol')['Close'].transform(lambda x: x.ewm(span = 12).mean())
macd = ema_12-ema_26

#calculate the EMA 
ema_9_macd = macd.ewm(span = 9).mean()
 
# store the data in data frame 
df['MACD']= macd
df['MACD_EMA'] = ema_9_macd
```

```python
df.head(30)
```

```python
# calculate price rate of change 
n=9

#calculate the rate of change in the price and store it in the data frame
df['PROC'] = df.groupby('Symbol')['Close'].transform(lambda x: x.pct_change(periods = n))
```

```python
df.head(30)
```

```python
def obv(group):
  volume = group['Volume']
  change = group['Close'].diff()

  prev_obv=0
  obv_values= []

  for i,j in zip(change, volume):

    if i>0:
      current_obv=prev_obv + j
    elif i<0:
      current_obv=prev_obv - j
    else:
      current_obv = prev_obv
    
    prev_obv = current_obv
    obv_values.append(current_obv)

  return pd.Series(obv_values, index =group.index)


obv_groups = df.groupby('Symbol').apply(obv)

df['on Balance volume'] = obv_groups.reset_index(level=0, drop = True)

df.head(30)
```
